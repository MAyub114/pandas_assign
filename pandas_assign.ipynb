# Better Pandas using Assign()

### Index assignment vs. .assign() method



```python
import numpy as np
import pandas as pd
```

### Load dataset

```python
FILE_PATH = '/workspaces/pandas_assign/data/'
ames = pd.read_csv(FILE_PATH + 'ames.csv')
print(f"Ames{ames.shape}")
ames.head()
```

### Add a new column

* Add a new column called "LogSalePrice" to the Ames dataset


#### Index assignment (bad approach) 

* Take a copy of the Ames dataset
* A copy is needed to avoid the "SettingWithCopyWarning" that plagues Pandas users

```python
ames_bad = ames.copy()
ames_bad['LogSalePrice'] = np.log(ames_bad.SalePrice)
ames_bad.head()
```

#### Assign method (good approach)

This is how to do the same using .assign()

```python
(ames
 .assign(LogSalePrice=np.log(ames.SalePrice))
 .head()
)
```
